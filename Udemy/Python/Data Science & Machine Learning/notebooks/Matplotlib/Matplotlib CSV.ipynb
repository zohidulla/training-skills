```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('jersey-rainfall-1984-to-2021.csv', index_col=0)
df.head()
```

```python
df.plot(y='YearTotalmm', kind='area')
plt.show()
```

```python
df.plot(y='Sep')
plt.show()
```

```python
df.plot(y=['Apr', 'May'], kind='area')
plt.show()
```

```python
df.plot()
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).plot()
plt.show()
```
